```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

def calculate_bmi(weight, height_m):
    return weight / (height_m ** 2)

def classify_bmi(bmi):
    if bmi < 18.5:
        return "Underweight", "blue"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight", "green"
    elif 25 <= bmi < 29.9:
        return "Overweight", "orange"
    else:
        return "Obese", "red"

weight = float(input("Enter weight (kg): "))
height_cm = float(input("Enter height (cm): "))
height_m = height_cm / 100 

bmi = calculate_bmi(weight, height_m)
category, color = classify_bmi(bmi)

print(f"\nYour BMI is: {bmi:.2f}")
print(f"Category: {category}")

heights = np.linspace(140, 200, 50)
weights = np.linspace(40, 120, 50)

H, W = np.meshgrid(heights / 100, weights)
BMI_values = W / (H ** 2)

levels = [0, 18.5, 24.9, 29.9, 35]  
colors = ["blue", "green", "orange", "red"]
labels = ["Underweight (<18.5)", "Normal weight (18.5 - 24.9)", 
          "Overweight (25 - 29.9)", "Obese (30+)"]

plt.figure(figsize=(8, 6))
plt.contourf(H * 100, W, BMI_values, levels=levels, colors=colors, alpha=0.7)
plt.colorbar(label="BMI Value")

plt.scatter(height_cm, weight, color=color, edgecolors="black", s=100, label="Your BMI")

plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("BMI Classification Chart")
plt.grid(True)

legend_patches = [Patch(color=colors[i], label=labels[i]) for i in range(len(labels))]
plt.legend(handles=legend_patches, loc="upper right")

plt.show()
```
